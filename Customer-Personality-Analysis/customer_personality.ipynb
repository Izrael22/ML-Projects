### DATA PREPROCESSING

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
scaled_df = pd.DataFrame(scaler.transform(df),columns= df.columns )
```

```python
scaled_df.head()
```

##### Dimensionality reduction with PCA

```python
from sklearn.decomposition import PCA
```

```python
model = PCA(n_components=3)
model.fit(scaled_df)
model_df = pd.DataFrame(model.transform(scaled_df), columns=(["col1","col2", "col3"]))
model_df.describe().T
```

### Elbow Method

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(model_df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('the number of clusters')
plt.ylabel('wcss')
plt.show()
```

#### As it is not very clear from the elbow method that which value of K to choose

### Silhouette Score

```python
from sklearn.metrics import silhouette_score 
```

```python
silhouette_scores = []
for i in range(2,10):
    m1=KMeans(n_clusters=i, random_state=42)
    c = m1.fit_predict(model_df)
    silhouette_scores.append(silhouette_score(model_df, m1.fit_predict(scaled_df))) 
plt.bar(range(2,10), silhouette_scores) 
plt.xlabel('Number of clusters', fontsize = 20) 
plt.ylabel('S(i)', fontsize = 20) 
plt.show()
```

```python
silhouette_scores
```

#### Getting the maximum value of silhouette score and adding 2 in index because index starts from 2.

```python
sc=max(silhouette_scores)
number_of_clusters=silhouette_scores.index(sc)+2
print("Number of Cluster Required is : ", number_of_clusters)
```

```python
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=0)
y_kmeans = kmeans.fit_predict(scaled_df)
```

```python
df['cluster']= y_kmeans + 1
```
